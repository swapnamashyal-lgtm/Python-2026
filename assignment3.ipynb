```python


item_name = input("Enter item name: ")
price = float(input("Enter item price: "))
quantity = int(input("Enter quantity: "))

total = price * quantity

print("\n--- Order Summary ---")
print("Item:", item_name)
print("Price per item:", price)
print("Quantity:", quantity)
print("Total cost:", total)
print("---------------------")
```

```python


visitors = int(input("Enter number of visitors today: "))
page_views = int(input("Enter total page views: "))

average = page_views / visitors

print("\n--- Website Traffic Report ---")
print("Visitors:", visitors)
print("Page Views:", page_views)
print("Average pages per visitor:", average)
```

```python


distance = int(input("Enter delivery distance in km: "))

if distance <= 5:
    print("Category: Near")
elif distance <= 15:
    print("Category: Medium")
else:
    print("Category: Far")
```

```python


source = input("Enter exam source (Mobile/Laptop/Desktop): ")

if source == "Mobile":
    print("Exam taken from Mobile device")
elif source == "Laptop":
    print("Exam taken from Laptop")
elif source == "Desktop":
    print("Exam taken from Desktop computer")
else:
    print("Unknown exam source")
```

```python


feedbacks = []
for i in range(3):
    fb = input(f"Enter feedback {i+1}: ")
    feedbacks.append(fb)

print("\n--- Feedback Dashboard ---")
for i, fb in enumerate(feedbacks, start=1):
    print(f"{i}. {fb}")
```

```python
contacts = {}
def add_contact():
    name = input("Name: ")
    phone = input("Phone: ")
    contacts[name] = phone
def search_contact():
    name = input("Name: ")
    print(contacts.get(name, "Not found"))
def delete_contact():
    name = input("Name: ")
    contacts.pop(name, None)
while True:
    print("1.Add 2.Search 3.Delete 4.Exit")
    ch = input("Choice: ")
    if ch == "1":
        add_contact()
    elif ch == "2":
        search_contact()
    elif ch == "3":
        delete_contact()
    elif ch == "4":
        break
```

```python
tasks = []
def add_task():
    tasks.append(input("Task: "))
def view_tasks():
    for i, t in enumerate(tasks, 1):
        print(i, t)
def remove_task():
    view_tasks()
    idx = int(input("Task no: "))
    if idx <= len(tasks):
        tasks.pop(idx-1)
while True:
    print("1.Add 2.View 3.Remove 4.Exit")
    ch = input("Choice: ")
    if ch == "1":
        add_task()
    elif ch == "2":
        view_tasks()
    elif ch == "3":
        remove_task()
    elif ch == "4":
        break
```

```python
balance = 10000
def check_balance():
    print("Balance:", balance)
def deposit():
    global balance
    balance += int(input("Amount: "))
def withdraw():
    global balance
    amt = int(input("Amount: "))
    if amt <= balance:
        balance -= amt
while True:
    print("1.Balance 2.Deposit 3.Withdraw 4.Exit")
    ch = input("Choice: ")
    if ch == "1":
        check_balance()
    elif ch == "2":
        deposit()
    elif ch == "3":
        withdraw()
    elif ch == "4":
        break
```

```python
questions = [
 {"q":"Capital of India?", "a":"Delhi"},
 {"q":"2+3?", "a":"5"}
]
score = 0
for q in questions:
    if input(q["q"]+" ").lower() == q["a"].lower():
        score += 1
print("Score:", score)
```

```python
expenses = {}
def add_expense():
    cat = input("Category: ")
    amt = int(input("Amount: "))
    expenses[cat] = expenses.get(cat, 0) + amt
def view_expenses():
    for c,a in expenses.items():
        print(c, a)
while True:
    print("1.Add 2.View 3.Exit")
    ch = input("Choice: ")
    if ch == "1":
        add_expense()
    elif ch == "2":
        view_expenses()
    elif ch == "3":
        break
```
